```python
import cv2
from google.colab.patches import cv2_imshow


while True:
  a=int(input('フィルターの番号を選択　　'))
  if a==0 or a==1 or a==2 or a==3:
    break

print('フィルターにかける画像ファイルをアップロードしてください')
f = input('ファイル名を入力(拡張子を除く)')
if a == 0:
  image = cv2.imread(f+'.jpg')

  kernel_size = 5

  blurred = cv2.blur(image, (kernel_size, kernel_size))

  cv2_imshow(blurred)

elif a==1:
  import cv2

  image = cv2.imread(f+'.jpg')

  kernel_size = 5

  sigma = 0

  blurred = cv2.GaussianBlur(image, (kernel_size, kernel_size), sigma)

  cv2_imshow(blurred)


elif a==2:
  import cv2

  image = cv2.imread(f+'.jpg')

  kernel_size = 5

  blurred = cv2.medianBlur(image, kernel_size)

  cv2_imshow(blurred)

elif a==3:
  import cv2

  image = cv2.imread(f+'.jpg')

  d = 5

  sigmaColor = 75

  sigmaSpace = 75

  blurred = cv2.bilateralFilter(image, d, sigmaColor, sigmaSpace)

  cv2_imshow(blurred)
```
